## 2) Hypothesis testing  : test if two groups are different ?
Select the right test for the right data. 
### A) Normal distributions

- Welch test (and ignore the T-test): Fewer assumptions for the Welch test vs. the T-test because it perform poorly

### B) Non normal distributions

- Wilcoxon-test: A good test for _non normal_ data with very few assumptions. We are testing the hypothesis that two datasets have different distributions. The main assumptions here are: 1) samples are randomly representative of population, 2) samples are independent of each other , 3) values have an order (e.g. 3 is more than 1, but we canâ€™t say that true is more than false).

> it is important to calculate the power (which is related to sample size) to make sure our study has sufficient power. Remember, low power means we are unlikely to detect an effect from an intervention, and generally makes our results less reliable.
